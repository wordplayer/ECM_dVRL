```python
import gym
import time
import numpy as np
import ECM_dVRL_v01
import numdifftools as nd
```

```python
env = gym.make('ECM-v0')
```

```python
env.reset()
```

```python
env.get_achieved_goal()
```

```python
theta = env.ecm.getJointAngles()
lr = 1e-4
theta_new = 0.
tolerance = 1e-3
max_iter = 100
iterations = 0
while iterations < max_iter:
    x_diff = env.get_achieved_goal()[:-1] - env.desired_goal[:-1]
    J = env.ecm.getJacobian().T
    theta_new = lr * np.dot(np.linalg.pinv(J), x_diff) + theta
    if np.linalg.norm(theta_new - theta) < tolerance:
        print (np.linalg.norm(theta_new - theta), theta_new)
        break
    else:
        print (np.linalg.norm(theta_new - theta), theta_new)
    theta = theta_new
    iterations += 1
    time.sleep(1.0)
```

```python
env.close()
```
