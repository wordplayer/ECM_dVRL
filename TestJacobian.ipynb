```python
import gym
import time
import numpy as np
import ECM_dVRL_v01
import numdifftools as nd
```

```python
env = gym.make('ECM-v0')
```

```python
env.reset()
```

```python
env.get_achieved_goal()
```

```python
env.ecm.getTestGoalPosition()
```

```python
theta = env.ecm.getJointAngles()
lr = 1e-4
theta_new = 0.
tolerance = 1e-3
max_iter = 100
iterations = 0
while True:
    x_diff = env.get_achieved_goal()[:-1] - env.desired_goal[:-1]
    J = env.ecm.getJacobian().T
    theta_new = lr * np.dot(np.linalg.pinv(J), x_diff) + theta
    if np.linalg.norm(theta - theta_new) < tolerance:
        print (env.get_achieved_goal()[:-1], env.desired_goal[:-1], iterations)
        break
    else:
        print (env.get_achieved_goal()[:-1], theta, theta_new)
        env.ecm.setJointAngles(theta_new)
    theta = theta_new
    iterations += 1
    time.sleep(1.0)
```

```python
env.close()
```
